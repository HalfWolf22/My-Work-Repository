### Proven Influencers
Normally, various researches have proven that 4 factors influence the outcome of an NBA the most (Personal fauls, Turnovers, Defensive rebounds and Field goal percentage), as shown at the barplots in the cell down. Contrary to that statement, we don't have any of that information because we want to predict the outcome of a game before it even began thus we can't use stats that get produced during the game. Instead we are going to have to rely on "form" stats.

```python
fig, ax = plt.subplots(4, 1, figsize=(10, 28))

outlier_pf_teams = merged.loc[(merged['PF_team']<38)&(6<merged['PF_team'])]
sns.barplot(x='PF_team', y='Won', data=outlier_pf_teams, ax=ax[0])
ax[0].set_title('Influence of number of fauls to the outcome of the game',
                fontsize=14)
ax[0].text(25, 1, f"Correlation: {round(merged['PF_team'].corr(merged['Won']), 3)}")

           
outlier_tov_teams = merged.loc[merged['TOV_team']<27]
sns.barplot(x='TOV_team', y='Won', data=outlier_tov_teams, ax=ax[1])
ax[1].set_title('Influence of number of turnovers to the outcome of the game',
                fontsize=14)
ax[1].text(20, 1, f"Correlation: {round(merged['TOV_team'].corr(merged['Won']), 3)}")


outlier_dreb_teams = merged.loc[(merged['DREB_team']<38)&(15<merged['DREB_team'])]
sns.barplot(x='DREB_team', y='Won', data=outlier_dreb_teams, ax=ax[2])
ax[2].set_title('Influence of number of defensive rebounds to the outcome of the game',
                fontsize=14)
ax[2].text(17.8, 0.73, f"Correlation: {round(merged['DREB_team'].corr(merged['Won']), 3)}")


binned_fg = pd.qcut(merged['FG%_team'], 10)
sns.barplot(x=binned_fg, y=merged['Won'], ax=ax[3])
ax[3].set_title('Influence of field goal percentage to the outcome of the game',
                fontsize=14)
ax[3].set_xticklabels(rotation=45, labels=binned_fg.unique().sort_values())
ax[3].text(7.8, 0.8, f"Correlation: {round(merged['FG%_team'].corr(merged['Won']), 3)}")
```

# Individual feature inspection

```python
import random

print('---------------------\nHighly skewed, right tail:\n---------------------\n')
high_right = []
for feature in data.columns:
    if data[feature].skew()>1:
        high_right.append(feature)

print('Number of highly skewed, right tailed features: ', len(high_right))

mean = data[high_right].mean()
imputed_mean = pd.Series({feature: data[feature][data[feature]!=0].skew() for feature in high_right})
diff = round((mean-imputed_mean).mean(), 3)
print('Skewness mean - skewness mean with trimmed zeroes mean difference:', diff, '\n')
        
for feature in random.sample(high_right, 5):
    print(feature, data[feature].skew().round(3))
    eds.plot_spread(data, feature)
        
print('---------------------\nHighly skewed, left tail:\n---------------------\n')
for feature in data.columns:
    if data[feature].skew()<-1:
        print(feature, data[feature].skew().round(3))
        eds.plot_spread(data, feature)
```

### Highly skewed features
It seems that the 0 imputation messes up several same types of columns (AST, BLK, REB and 3PM for right and percentages for left tail) which should actaully be expected given the high missing percentages (~15%-20%). For the left skew cases, there shouldn't be many problems since they really look like an arbitrary values that are going to be nicely interpreted by a tree based algorithm. 

Right tailed ones are a bit trickier. As the difference result shows, if we trim the zero entries, the skewness gets even worse overall! The problem is that that data is naturally scewed, since we are getting forms, a 20% presece of zeroes in data results in many values close to zero and that is why it is scewed. Same for the left, FT% goes around 80% average over the league and this many zeroes make percentage columns skewed.

```python
print('---------------------\nModerately skewed, right tail:\n---------------------\n')
mod_right = []
for feature in data.columns:
    if 0.7<data[feature].skew()<1:
        mod_right.append(feature)

print('Number of moderately skewed, right tailed features: ', len(mod_right), '\n')

for feature in random.sample(mod_right, 5):
    print(feature, data[feature].skew().round(3))
    eds.plot_spread(data, feature)
    
print('---------------------\nModerately skewed, left tail:\n---------------------\n')
mod_left = []
for feature in data.columns:
    if -0.7>data[feature].skew()>-1:
        mod_left.append(feature)
        
print('Number of moderately skewed, left tailed features: ', len(mod_left), '\n')
        
for feature in random.sample(mod_left, 5):
    print(feature, data[feature].skew().round(3))
    eds.plot_spread(data, feature)
```

### Moderately skewed features
Similar situation can be spotted in moderately skewed features. Right tailed features are just features with similar characteristics as highly skewed right tailed features. STL are in quantity a very similar feature with BLK, but BLKs tend to have higher extremes in game and thus are more skewed.

When it comes to left tailed skewed features, it is pretty much the same, only here are less important players more present since they have to be more constant on lower possesions (a bit less likely to have lows and are less likely to have extreme highs).

```python
print('---------------------\nLow skeweness, right tail:\n---------------------\n')
low_right = []
for feature in data.columns:
    if 0.5<data[feature].skew()<0.7:
        low_right.append(feature)

print('Number of lowly skewed, right tailed features: ', len(low_right), '\n')

for feature in random.sample(low_right, 5):
    print(feature, data[feature].skew().round(3))
    eds.plot_spread(data, feature)
    
print('---------------------\nLow skewness, left tail:\n---------------------\n')
low_left = []
for feature in data.columns:
    if -0.5>data[feature].skew()>-0.7:
        low_left.append(feature)
        
print('Number of lowly skewed, left tailed features: ', len(low_left), '\n')
        
for feature in random.sample(low_left, 5):
    print(feature, data[feature].skew().round(3))
    eds.plot_spread(data, feature)
```

### Lowly skewed features
Most of the right tailed ones looks similar to the moderately skewed features. The others don't seem to pose a problem at least visually, as I said, I will only deal with the outliers if models don't yield satisfactory results.

The nulls at the left skewed ones nicely represent the nulls as arbitrary values. The only exepcion is Season_Half_is2 where there are more 1s than 0s and that is logical since NBA regular season is played for only 3 months in the first half of a year (0) and 4 months in the second half (1).

# Difference data

```python
aways = data.iloc[:, 4:116]
homes = data.iloc[:, 116:-2]
diff_data = pd.DataFrame()
for col in aways.columns:
    diff_data[col[:-5]] = aways[col]-homes[col[:-5]+'_home']
diff_data = pd.concat([data.iloc[:, :4], data.iloc[:, -2:], diff_data], axis=1)

print('---------------------\nCorrelation of difference data features with the target:')
print('---------------------')
eds.print_abv_corr(diff_data.drop(['WonH1_away', 'WonQ1_away'], axis=1),
                   diff_data['Won_away'], 0.07)

print('\n---------------------\nForrest correlation of difference data features with the target:')
print('---------------------')
eds.check_forest_corr(diff_data.drop(['Won_away', 'WonH1_away', 'WonQ1_away'], axis=1), 
                      diff_data['Won_away'], 'binary:logistic', threshold=0.07)
```

### Diference data
Ever since I started planing this NBA project, I had this varition of the dataset in mind I would use for predictions (and it seems that my intuition isn't failing me). Extracting the difference between the corresponding stats makes features more rigorous on their own so, in the end, 2 useless features are merged and the new one is created that might be useful.

I can't say right now which of these two datasets is going to perform better prediciton-wise so I am going to wait for that, but till then, I can still analyse this new data.

```python
print('---------------------\nSkewed features:\n---------------------')
for feature in diff_data.columns:
    if abs(diff_data[feature].skew())>0.5:
        print(feature, diff_data[feature].skew().round(3))
print('\nDone!\n')

for feature in random.sample(list(diff_data.columns), 5):
    print(feature, diff_data[feature].skew().round(3))
    eds.plot_spread(diff_data, feature)
```

### Skewness of the difference data
Great! The subtraction of the features centers them around zero and gives them pretty much Gaussian-like look. That can be of an extreme use since the data requires little to no processing (for linear models, as logistic regression for example).
